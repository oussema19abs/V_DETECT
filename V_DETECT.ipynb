```python
import os

def print_custom_tree(startpath, file_display_limit=10):
    """
    Prints a directory tree structure with file counts for directories
    exceeding a certain file_display_limit.
    """
    total_dirs = 0
    total_files = 0

    for root, dirs, files in os.walk(startpath):
        total_dirs += 1
        total_files += len(files)

        level = root.replace(startpath, '').count(os.sep)

        if root == startpath:
            indent = ''
            dir_name_prefix = ''
        else:
            indent = '│   ' * level
            dir_name_prefix = '├── '

        print(f"{indent}{dir_name_prefix}{os.path.basename(root)}/")

        if files:
            if len(files) > file_display_limit:
                content_indent = '│   ' * (level + 1)
                print(f"{content_indent}└── [ {len(files)} files ]")
            else:
                content_indent = '│   ' * (level + 1)
                for i, f in enumerate(sorted(files)):
                    file_prefix = '├── ' if i < len(files) - 1 else '└── '
                    print(f"{content_indent}{file_prefix}{f}")

    print(f"\n{total_dirs} directories, {total_files} files (Some file listings summarized).")


print("--- Full Directory Tree of /content/ (with file counts for large folders) ---")
print_custom_tree('/content/')
```

```python
import os
import cv2
import random
import numpy as np
from tqdm import tqdm
import tensorflow as tf

from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input,
    TimeDistributed,
    GlobalAveragePooling2D,
    Bidirectional,
    GRU,
    Dense,
    Dropout
)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, TensorBoard
from tensorflow.keras.applications import MobileNetV2

from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report, confusion_matrix

IMG_SIZE = 112
SEQUENCE_LENGTH = 60
BATCH_SIZE = 16
INITIAL_EPOCHS = 20
FINE_TUNE_EPOCHS = 30

DATA_DIR = '/content/Combined_Dataset'
MODEL_PATH = '/content/fight_detection_poc_model.h5'

print(f"Constants are set. Main data directory is now: {DATA_DIR}")
print(f"Image size: {IMG_SIZE}x{IMG_SIZE}, Sequence length: {SEQUENCE_LENGTH}, Batch size: {BATCH_SIZE}")
```

```python
def preprocess_video(video_path, augment=False):
    """
    Loads a video, efficiently samples frames by reading sequentially,
    resizes them, and normalizes pixel values. Applies optional augmentation.

    Args:
        video_path (str): The full path to the video file.
        augment (bool): Whether to apply data augmentation (e.g., horizontal flip).

    Returns:
        np.array: A NumPy array of preprocessed frames with shape
                  (SEQUENCE_LENGTH, IMG_SIZE, IMG_SIZE, 3) and dtype np.float32,
                  or None if processing fails.
    """
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return None

    frames = []
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if total_frames == 0:
        cap.release()
        return None

    if total_frames < SEQUENCE_LENGTH:
        frame_indices = np.arange(total_frames)
    else:
        frame_indices = np.linspace(0, total_frames - 1, SEQUENCE_LENGTH).astype(int)

    ret = True
    current_frame_pos = 0
    target_frame_idx = 0

    while ret and target_frame_idx < len(frame_indices):
        ret, frame = cap.read()

        if ret and current_frame_pos == frame_indices[target_frame_idx]:
            try:
                frame_resized = cv2.resize(frame, (IMG_SIZE, IMG_SIZE))
                frame_normalized = frame_resized / 255.0

                if augment and np.random.rand() > 0.5:
                    frame_normalized = np.fliplr(frame_normalized)

                frames.append(frame_normalized.astype(np.float32))
                target_frame_idx += 1
            except Exception as e:
                print(f"WARNING: Error processing frame {current_frame_pos} from {video_path}: {e}")

        current_frame_pos += 1

    cap.release()

    while len(frames) < SEQUENCE_LENGTH:
        frames.append(np.zeros((IMG_SIZE, IMG_SIZE, 3), dtype=np.float32))

    if len(frames) == SEQUENCE_LENGTH:
        return np.array(frames, dtype=np.float32)
    else:
        print(f"CRITICAL WARNING: Processed video {video_path} resulted in {len(frames)} frames, expected {SEQUENCE_LENGTH}.")
        return None


print("`preprocess_video` function defined with optimized frame reading.")
```

```python
def create_dataset_from_folder(folder_path, is_training=False):
    """
    Creates a tf.data.Dataset from video files in the specified folder.
    This dataset will stream and preprocess videos efficiently.

    Args:
        folder_path (str): The base directory containing 'Fight' and 'NonFight' subfolders.
        is_training (bool): If True, applies data augmentation and shuffling.

    Returns:
        tuple: A tuple containing (tf.data.Dataset, list_of_original_labels).
               The list_of_original_labels is used for class weight computation.
    """
    video_paths = []
    labels = []

    for label_name in ['Fight', 'NonFight']:
        class_path = os.path.join(folder_path, label_name)

        if not os.path.exists(class_path):
            print(f"WARNING: Class path does not exist: {class_path}. Skipping {label_name}.")
            continue

        current_video_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.mp4', '.avi', '.mpg', '.mov', '.mkv'))]

        if not current_video_files:
            print(f"WARNING: No video files found in {class_path}. Skipping {label_name}.")
            continue

        for video_file in current_video_files:
            full_path = os.path.join(class_path, video_file)
            video_paths.append(full_path)
            labels.append(1 if label_name == 'Fight' else 0)

    if not video_paths:
        print(f"ERROR: No video files found in {folder_path}. Returning empty dataset.")
        empty_dataset = tf.data.Dataset.from_tensor_slices((tf.constant([], dtype=tf.string), tf.constant([], dtype=tf.int32)))
        empty_dataset = empty_dataset.map(lambda x, y: (
            tf.zeros((SEQUENCE_LENGTH, IMG_SIZE, IMG_SIZE, 3), dtype=tf.float32),
            tf.constant(0, dtype=tf.int32)
        ))
        empty_dataset = empty_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
        return empty_dataset, []

    dataset_paths = tf.constant(video_paths, dtype=tf.string)
    dataset_labels = tf.constant(labels, dtype=tf.int32)

    dataset = tf.data.Dataset.from_tensor_slices((dataset_paths, dataset_labels))

    def _preprocess_video_tf(video_path_tensor, label_tensor):
        video_path = video_path_tensor.numpy().decode('utf-8')
        processed_video = preprocess_video(video_path, augment=is_training)

        if processed_video is None:
            return tf.constant(0, dtype=tf.float32), tf.constant(-99, dtype=tf.int32)

        return tf.convert_to_tensor(processed_video, dtype=tf.float32), label_tensor

    dataset = dataset.map(
        lambda x, y: tf.py_function(
            _preprocess_video_tf,
            inp=[x, y],
            Tout=[tf.float32, tf.int32]
        ),
        num_parallel_calls=tf.data.AUTOTUNE
    )

    dataset = dataset.filter(lambda x, y: y != -99)

    dataset = dataset.map(lambda x, y: (
        tf.ensure_shape(x, (SEQUENCE_LENGTH, IMG_SIZE, IMG_SIZE, 3)),
        tf.ensure_shape(y, ())
    ))

    if is_training:
        # Shuffle the dataset.
        dataset = dataset.shuffle(buffer_size=min(len(video_paths), 1024))
        dataset = dataset.repeat()

    if not is_training:
        dataset = dataset.repeat()


    # Batch the dataset.
    dataset = dataset.batch(BATCH_SIZE)

    dataset = dataset.prefetch(tf.data.AUTOTUNE)

    return dataset, labels

print("`create_dataset_from_folder` function defined with shape correction and `.repeat()`.")
```

```python
print("Creating training dataset...")
train_dataset, train_labels_list = create_dataset_from_folder(os.path.join(DATA_DIR, 'train'), is_training=True)

print("\nCreating validation dataset...")
val_dataset, val_labels_list = create_dataset_from_folder(os.path.join(DATA_DIR, 'val'), is_training=False)


print("\nVerifying a batch from training dataset:")
try:
    for X_batch_train, y_batch_train in train_dataset.take(1):
        print(f"Training batch data shape: {X_batch_train.shape}, dtype: {X_batch_train.dtype}")
        print(f"Training batch labels shape: {y_batch_train.shape}, dtype: {y_batch_train.dtype}")
        if np.isnan(X_batch_train.numpy()).any() or np.isinf(X_batch_train.numpy()).any():
            print("CRITICAL WARNING: Training batch contains NaN or Inf values!")
except tf.errors.OutOfRangeError:
    print("Training dataset is empty.")
except Exception as e:
    print(f"Error inspecting training dataset: {e}")

print("\nVerifying a batch from validation dataset:")
try:
    for X_batch_val, y_batch_val in val_dataset.take(1):
        print(f"Validation batch data shape: {X_batch_val.shape}, dtype: {X_batch_val.dtype}")
        print(f"Validation batch labels shape: {y_batch_val.shape}, dtype: {y_batch_val.dtype}")
        if np.isnan(X_batch_val.numpy()).any() or np.isinf(X_batch_val.numpy()).any():
            print("CRITICAL WARNING: Validation batch contains NaN or Inf values!")
except tf.errors.OutOfRangeError:
    print("Validation dataset is empty.")
except Exception as e:
    print(f"Error inspecting validation dataset: {e}")

if not train_labels_list:
    print("\nFATAL ERROR: No training data found. Training cannot proceed.")
if not val_labels_list:
    print("\nFATAL ERROR: No validation data found. Training cannot proceed.")
```

```python
def build_transfer_learning_model(input_shape=(SEQUENCE_LENGTH, IMG_SIZE, IMG_SIZE, 3)):
    base_model = MobileNetV2(input_shape=(IMG_SIZE, IMG_SIZE, 3),
                             include_top=False,
                             weights='imagenet')

    for layer in base_model.layers[:100]: # Freezes first 100 out of ~154 layers
        layer.trainable = False

    # Defines the input layer for video sequences
    inputs = Input(shape=input_shape)

    # Applies the base_model (feature extractor) to each frame in the sequence
    x = TimeDistributed(base_model)(inputs)
    # Applies GlobalAveragePooling2D to the output of MobileNetV2 for each frame
    x = TimeDistributed(GlobalAveragePooling2D())(x)

    x = Bidirectional(GRU(64, return_sequences=True))(x)
    x = GRU(128)(x)

    x = Dense(64, activation='relu')(x)
    x = Dropout(0.3)(x)
    outputs = Dense(1, activation='sigmoid')(x)

    return Model(inputs, outputs)

model = build_transfer_learning_model()
model.summary()

print("\nModel built successfully.")
```

```python
optimizer = Adam(learning_rate=1e-4, weight_decay=1e-6)

if train_labels_list:
    class_weights = compute_class_weight('balanced',
                                       classes=np.unique(train_labels_list),
                                       y=train_labels_list)
    class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}
    print(f"Computed class weights: {class_weight_dict}")
else:
    print("WARNING: No training labels available. Skipping class weight computation. This might indicate a data loading issue.")
    class_weight_dict = None

callbacks = [
    # Stops training if validation loss doesn't improve for 'patience' epochs
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    # Reduces learning rate when validation loss stops improving
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1),
    # Saves the best model weights based on validation loss
    ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True, verbose=1),
    # For visualizing training metrics
    TensorBoard(log_dir='./logs')
]

model.compile(optimizer=optimizer,
              loss='binary_crossentropy',
              metrics=['accuracy'])

if train_labels_list and val_labels_list:
    print("\nStarting initial model training...")
    num_train_videos = len(train_labels_list)
    num_val_videos = len(val_labels_list)
    steps_per_epoch = num_train_videos // BATCH_SIZE
    validation_steps = num_val_videos // BATCH_SIZE

    print(f"\nStarting initial model training with {steps_per_epoch} steps per epoch...")
    history = model.fit(
        train_dataset,
        validation_data=val_dataset,
        epochs=INITIAL_EPOCHS,
        class_weight=class_weight_dict,
        callbacks=callbacks,
        steps_per_epoch=steps_per_epoch,
        validation_steps=validation_steps
    )
    print("\nInitial model training completed.")
else:
    print("\nCannot start initial training: Training or validation data is empty. Please check data loading steps and error messages above.")
```

```python
print("\n--- PHASE 2: FINE-TUNING THE ENTIRE MODEL ---")

model.layers[1].trainable = True
model.layers[1].layer.trainable = True # Unfreezes the entire MobileNetV2 base model

fine_tune_at = 140 # Unfreezes layers from this index onwards in MobileNetV2
for layer in model.layers[1].layer.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(optimizer=Adam(learning_rate=1e-5),
              loss='binary_crossentropy',
              metrics=['accuracy'])

print(f"Re-compiling model for fine-tuning. {len(model.trainable_variables)} trainable variables.")
model.summary()

fine_tune_callbacks = [
    EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', patience=2, factor=0.2, verbose=1),
    ModelCheckpoint('best_model_finetuned.h5', monitor='val_loss', save_best_only=True, verbose=1),
    TensorBoard(log_dir='./logs/fine_tune')
]

total_epochs = INITIAL_EPOCHS + FINE_TUNE_EPOCHS
if train_labels_list and val_labels_list:
    num_train_videos = len(train_labels_list)
    num_val_videos = len(val_labels_list)
    steps_per_epoch = num_train_videos // BATCH_SIZE
    validation_steps = num_val_videos // BATCH_SIZE
    print("\nStarting fine-tuning...")
    history_fine_tune = model.fit(
        train_dataset,
        validation_data=val_dataset,
        epochs=total_epochs,
        initial_epoch=history.epoch[-1] + 1,
        callbacks=fine_tune_callbacks,
        steps_per_epoch=steps_per_epoch,
        validation_steps=validation_steps
    )

    print("✅ Training complete. Saving the final model...")
    model.save(MODEL_PATH)
    print(f"Model saved to '{MODEL_PATH}'")
else:
    print("\nSkipping fine-tuning: Training or validation data is empty.")
```

```python
def preprocess_for_prediction(video_path):
    """
    Preprocesses a single video for prediction.
    This is a robust version, similar to preprocess_video but without augmentation.
    Returns a NumPy array of frames (SEQUENCE_LENGTH, IMG_SIZE, IMG_SIZE, 3) or None on failure.
    """
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"ERROR: Could not open video file for prediction: {video_path}. Skipping.")
        return None

    final_frames = []
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if total_frames == 0:
        print(f"WARNING: Video {video_path} has 0 frames for prediction. Skipping.")
        cap.release()
        return None
    elif total_frames < SEQUENCE_LENGTH:
        frame_indices = np.linspace(0, total_frames - 1, total_frames).astype(int)
    else:
        frame_indices = np.linspace(0, total_frames - 1, SEQUENCE_LENGTH).astype(int)

    current_frame_count = 0
    frame_indices = sorted(list(set(frame_indices)))

    for i in frame_indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, i)
        ret, frame = cap.read()
        if not ret or frame is None:
            continue

        try:
            frame = cv2.resize(frame, (IMG_SIZE, IMG_SIZE))
            frame = frame / 255.0
            final_frames.append(frame.astype(np.float32))
            current_frame_count += 1
        except Exception as e:
            print(f"ERROR: Problem processing frame {i} from {video_path} during resize for prediction: {e}. Skipping this video.")
            cap.release()
            return None

    cap.release()

    if current_frame_count < SEQUENCE_LENGTH:
        while len(final_frames) < SEQUENCE_LENGTH:
            final_frames.append(np.zeros((IMG_SIZE, IMG_SIZE, 3), dtype=np.float32))

    np_final_frames = np.array(final_frames, dtype=np.float32)
    expected_shape = (SEQUENCE_LENGTH, IMG_SIZE, IMG_SIZE, 3)

    if np_final_frames.shape != expected_shape:
        print(f"ERROR: Processed video {video_path} has final shape {np_final_frames.shape}, expected {expected_shape}. Returning None for prediction.")
        return None

    return np_final_frames

print("`preprocess_for_prediction` function defined.")
```

```python
def robust_evaluation(model, video_paths_with_labels):
    """
    Evaluates model on a list of video paths with known true labels.
    video_paths_with_labels is expected to be a list of tuples: [(video_path, true_label_int), ...].
    It robustly handles video preprocessing and prediction errors.
    """
    y_pred = []
    y_true_filtered = []

    for path, true_label in tqdm(video_paths_with_labels, desc="Evaluating"):
        try:
            frames = preprocess_for_prediction(path)
            if frames is None:
                continue

            pred = model.predict(np.expand_dims(frames, 0), verbose=0)[0][0]

            y_pred.append(1 if pred > 0.5 else 0)
            y_true_filtered.append(true_label)

        except Exception as e:
            print(f"ERROR: Problem predicting for {path}: {e}. Skipping this video for evaluation.")

    if not y_pred:
        print("WARNING: No videos were successfully processed for evaluation. Cannot compute metrics.")
        return { 'accuracy': 0.0, 'precision': 0.0, 'recall': 0.0 }

    accuracy = accuracy_score(y_true_filtered, y_pred)
    precision = precision_score(y_true_filtered, y_pred, zero_division=0)
    recall = recall_score(y_true_filtered, y_pred, zero_division=0)

    print("\n" + "="*50)
    print("Evaluation Results")
    print("="*50)
    print(classification_report(y_true_filtered, y_pred, target_names=['NonFight', 'Fight'], zero_division=0))
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_true_filtered, y_pred))
    print("="*50)
    print(f"Final Accuracy: {accuracy:.2%}")
    print("="*50)

    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall
    }

print("`robust_evaluation` function defined.")
```
